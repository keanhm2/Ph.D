```python
from pyMetaheuristic.algorithm import whale_optimization_algorithm
from keras.datasets import mnist
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
import tensorflow as tf

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize and reshape the data
x_train = x_train / 255.0
x_test = x_test / 255.0
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Convert the labels to one-hot encoding #y_train 
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)

# Define CNN model function
def create_model(variables_values = [0, 0, 0, 0, 0, 0]):
    variables_values = variables_values.astype(int)
    filters, kernel_size_x, kernel_size_y, pool_size_x, pool_size_y, units = variables_values
    if filters==0:
      filters = 60
    if kernel_size_x ==0:
      kernel_size_x =3
    if kernel_size_y==0:
      kernel_size_y=4
    if pool_size_x ==0:
      pool_size_x =2
    if pool_size_y==0:
      pool_size_y=2
    if units==0:
      units = 69
      print("hata")

    print("filters: " + str(filters))
    print("kernel size of conv2d: (" + str(kernel_size_x) + "," + str(kernel_size_y) + ")")
    print("pool size of pooling2d: (" + str(pool_size_x) + "," + str(pool_size_y) + ")")
    print("units of dense layer: " + str(units))

    model = Sequential()
    model.add(Conv2D(filters=filters, kernel_size=(kernel_size_x, kernel_size_y), activation='relu', input_shape=(28, 28, 1)))
    model.add(MaxPooling2D(pool_size=(pool_size_x, pool_size_y)))
    model.add(Flatten())
    model.add(Dense(units, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Fit and evaluate the model
    model.fit(x_train, y_train, epochs=1, batch_size=32)
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    print("loss is : " + str(loss))
    return -loss

# Define search space
search_space = [
    (16, 64),  # filters
    (3, 5),    # kernel_size_x
    (3, 5),    # kernel_size_y
    (2, 4),    # pool_size_x
    (2, 4),    # pool_size_y
    (32, 256), # units
]

# Define WOA parameters
parameters = {
    'hunting_party': 10,
    'spiral_param': 0.5,
    'min_values': [a for a, _ in search_space],
    'max_values': [b for _, b in search_space],
    'iterations': 10,
    'verbose': True,
    'target_function': create_model,
}

# Optimize CNN hyperparameters using WOA
woa = whale_optimization_algorithm(**parameters)

# Print the best parameters
#print("Best Parameters: ", best_params)
```
