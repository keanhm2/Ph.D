```python
import numpy as np
from pyMetaheuristic.algorithm import whale_optimization_algorithm as WOA
from pyMetaheuristic.utils import graphs

# Target Function - It can be any function that needs to be minimize, However it has to have only one argument: 'variables_values'. This Argument must be a list of variables.
# For Instance, suppose that our Target Function is the Sphere Function (With two variables x1 and x2. Global Minimum f(x1, x2) = -1 for, x1 = 3.14 and x2 = 3.14)

# Target Function: Sphere Function
def sphere(variables_values = [0, 0]):
    x, y, z = variables_values
    func_value = (x - 1)**2 + (y - 2)**2 + (z - 3)**2
    return func_value

# WOA - Parameters
parameters = {
    'hunting_party': 150,
    'spiral_param': 0.5,
    'min_values': (-5, -5, -5),
    'max_values': (5, 5, 5),
    'iterations': 500,
    'verbose': True,
    'target_function': sphere
}

# WOA - Algorithm
#woa = whale_optimization_algorithm(hunting_party=150, spiral_param=0.5, min_values=(-5,-5,-5), max_values=(5,5,5), iterations=500, verbose=True, target_function=sphere)
woa = WOA(**parameters)
print(" ")
print(woa)
print(" ")
# WOA - Solution
variables = woa[0][:-1]
minimum   = woa[0][ -1]
print('Variables: ', np.around(variables, 4) , ' Minimum Value Found: ', round(minimum, 4) )

```
